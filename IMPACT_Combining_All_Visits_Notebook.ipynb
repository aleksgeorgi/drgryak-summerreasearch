# Merging baseline, visit 24, visit 26 datasets together

```python
import pandas as pd
import numpy as np
```

```python
impact_baseline = pd.read_excel("Data/IMPACT_Study/cleaned_datasets/impact_all_baseline_clean_final.xlsx")
impact_24 = pd.read_excel("Data/IMPACT_Study/cleaned_datasets/impact_all_24_clean_final.xlsx")
impact_26 = pd.read_excel("Data/IMPACT_Study/cleaned_datasets/impact_all_26_clean_final.xlsx")

```

```python
print(impact_baseline.shape)
print(len(set(impact_baseline['Participant ID'].unique())))
impact_baseline.head()
```

```python
print(impact_24.shape)
print(len(set(impact_24['Participant ID'].unique())))
impact_24.head()
```

```python
print(impact_26.shape)
print(len(set(impact_26['Participant ID'].unique())))
impact_26.head()
```

```python
# concatinating the DFs
impact_baseline_24 = pd.concat([impact_baseline, impact_24], ignore_index=True)
```

```python
print(impact_baseline_24.shape)
impact_baseline_24.head()
```

```python
impact_baseline_24_26 = pd.concat([impact_baseline_24, impact_26], ignore_index=True)
```

```python
print(impact_baseline_24_26.shape)
impact_baseline_24_26.head()
```

# Investigating why there are more unique participant IDs than we started with in the baseline dataset

```python
list_baseline = impact_baseline['Participant ID'].tolist()
list_24 = impact_24['Participant ID'].tolist()
list_26 = impact_26['Participant ID'].tolist()

common_participants = set(list_baseline).intersection(list_24, list_26)
common_participants = list(common_participants)
```

```python
len(set(common_participants))
```

```python
common_participants = list_baseline+list_24+list_26
common_participants = list(set(common_participants))
len(set(common_participants))
```

# Manually looking at original datasets shows that there are some participants who did not follow the chronological order of visits
Example: 
- some are missing visit 24 but have visit -2, 16, and 26
- 2 are missing baseline visits and did not get added to the baseline dataset

```python
# sorting dataset by participant ID and Visit

# Assuming you have a dataframe 'impact_baseline_24_26' and you want to sort it based on 'Participant ID' and 'Visit'
impact_combined_all = impact_baseline_24_26.sort_values(by=['Participant ID', 'Visit'], ascending=[True, True])

# Reset the index after sorting
impact_combined_all = impact_combined_all.reset_index(drop=True)

```
